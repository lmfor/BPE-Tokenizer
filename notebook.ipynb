```python
# find most common pair
def get_pair(ids):
    counts = {}

    # iterate consecutively
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1 #default to 0
    return counts

pair_stats = get_pair(tokens)
# print(pair_stats)
```

```python
def merge(ids, pair, idx):
    newids = []
    i = 0

    while i < len(ids):
        #    OB CHECK
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            newids.append(idx)
            i += 2
        else:
            newids.append(ids[i])
            i += 1

    return newids
```
